```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,1500)
mu = 428
sigma = 199
y = np.exp(-0.5 * ((x-mu)/sigma)**2) * ( 1 / (sigma * np.sqrt(2*np.pi)))
```

```python
plt.plot(x,y)
```

```python
factor = 10 / y[825] 
print(factor)
```

```python
y2 = np.exp(-0.5 * ((x-mu)/sigma)**2) * ( 1 / (sigma * np.sqrt(2*np.pi))) * factor
```

```python
plt.plot(x,y2)
```

```python
print(y2[825])
print(y2[29])
y2.max()
```

```python
FinalArea = np.trapz(y2[30:], x[30:], dx=0.000001)
```

```python
print('The total number of calculated sunbeds are:')
print(FinalArea)
```

```python
#################### try to find analytically the maximum point #######################
firstfactor = 1 / (sigma * np.sqrt(2 * np.pi)) * factor
print(firstfactor)
```

```python
################# alternative way ########################
trianglearea = (826 - 30) * (firstfactor - 10) /2
rectanglearea = (826 - 30) * 10

print('The total number of calculated sunbeds using triangle is:')
print(trianglearea + rectanglearea)
```
